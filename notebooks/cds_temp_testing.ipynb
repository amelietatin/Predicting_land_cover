```python
df_t=data_t['temperature_monthly-mean'].to_dataframe()
df_t
```

```python
df_t_reset = df_t.reset_index()
```

```python
import pandas as pd

df_t_reset['time'] = pd.to_datetime(df_t_reset['time'])

# Truncate the DataFrame to start from the year 2015
df_t_reset_truncated = df_t_reset[df_t_reset['time'].dt.year >= 2015]

# Reset the index if needed
df_t_reset_truncated.reset_index(inplace=True)
```

```python
df_t_reset_truncated
```

```python
# min/max coordinate values for sample = 1000
# min_lon= -31.169029141664772
# max_lon = 32.611141887507934
# min_lat = 27.746311186883922
# max_lat = 69.04164438989706
```

```python
# min/max coordinate values for sample = 10
min_lon= 4.875490655664919
max_lon = 27.392258798599347
min_lat = 35.4470142202193
max_lat = 62.138627784581054
```

```python
df_t_reset_truncated_filtered = df_t_reset_truncated[(df_t_reset_truncated["longitude"] >= min_lon) \
    & (df_t_reset_truncated["longitude"] <= max_lon) \
    & (df_t_reset_truncated["latitude"] >= min_lat) \
    & (df_t_reset_truncated["latitude"] <= max_lat)]
df_t_reset_truncated_filtered
```

```python
df_t_reset_truncated_filtered.set_index('index', inplace=True)
#df_t_reset_truncated_filtered.drop(columns=df_t_reset_truncated_filtered.columns[0], inplace=True)
df_t_reset_truncated_filtered
```
