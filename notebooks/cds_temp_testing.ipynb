```python
df_t=data_t['temperature_monthly-mean'].to_dataframe()
df_t
```

```python
df_t_reset = df_t.reset_index()
df_t_reset
```

```python
import pandas as pd

df_t_reset['time'] = pd.to_datetime(df_t_reset['time'])

# Truncate the DataFrame to start from the year 2015
df_t_reset_truncated = df_t_reset[df_t_reset['time'].dt.year >= 2015]

# Reset the index if needed
df_t_reset_truncated.reset_index(inplace=True)
```

```python
df_t_reset_truncated
```

```python
# min/max coordinate values for all PA
min_lon= -32.30906837949666
max_lon = 34.071555485388096
min_lat = 27.658764598717514
max_lat = 70.01981302035634
```

```python
df_t_filtered = df_t_reset_truncated[(df_t_reset_truncated["longitude"] >= min_lon) \
    & (df_t_reset_truncated["longitude"] <= max_lon) \
    & (df_t_reset_truncated["latitude"] >= min_lat) \
    & (df_t_reset_truncated["latitude"] <= max_lat)]
df_t_filtered
```

```python
df_t_filtered.set_index('index', inplace=True)
df_t_filtered
```

```python
import pandas as pd

# Assuming df_t_reset_truncated is your initial DataFrame
# Ensure the time column is in datetime format
df_t_filtered.loc[:, 'time'] = pd.to_datetime(df_t_filtered['time'])

# Create a new column for the quarter start date
df_t_filtered.loc[:, 'quarter_start'] = df_t_filtered['time'].dt.to_period('Q').dt.start_time

# Group by latitude, longitude, and the quarter start date
# Calculate the mean temperature for each group
df_t_quarterly = df_t_filtered.groupby(['latitude', 'longitude', 'quarter_start']).agg({
    'temperature_monthly-mean': 'mean'})

# Rename columns for clarity
df_t_quarterly.rename(columns={'temperature_monthly-mean': 'temperature_quarterly_mean'}, inplace=True)

#df_t_quarterly.set_index('index', inplace=True)
```
