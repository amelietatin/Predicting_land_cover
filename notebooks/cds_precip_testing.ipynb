```python
import xarray as xr

data_p= xr.open_dataset('../raw_data/precipitation_monthly-mean_bcc-csm1-1-m_rcp45_r1i1p1_1950-2100_v1.0.nc')
```

```python
df_p=data_p['precipitation_monthly-mean'].to_dataframe()
df_p
```

```python
df_p_reset = df_p.reset_index()
```

```python
import pandas as pd

df_p_reset['time'] = pd.to_datetime(df_p_reset['time'])

# Truncate the DataFrame to start from the year 2015
df_p_reset_truncated = df_p_reset[df_p_reset['time'].dt.year >= 2015]

# Reset the index if needed
df_p_reset_truncated.reset_index(inplace=True)
```

```python
df_p_reset_truncated
```

```python
# min/max coordinate values for sample = 1000

# min_lon= -31.169029141664772
# max_lon = 32.611141887507934
# min_lat = 27.746311186883922
# max_lat = 69.04164438989706
```

```python
# min/max coordinate values for all PA
min_lon= -32.30906837949666
max_lon = 34.071555485388096
min_lat = 27.658764598717514
max_lat = 70.01981302035634
```

```python
df_p_filtered = df_p_reset_truncated[(df_p_reset_truncated["longitude"] >= min_lon) \
    & (df_p_reset_truncated["longitude"] <= max_lon) \
    & (df_p_reset_truncated["latitude"] >= min_lat) \
    & (df_p_reset_truncated["latitude"] <= max_lat)]
df_p_filtered
```

```python
df_p_filtered.set_index('index', inplace=True)
df_p_filtered
```

```python
import pandas as pd

# Assuming df_t_reset_truncated is your initial DataFrame
# Ensure the time column is in datetime format
df_p_filtered.loc[:, 'time'] = pd.to_datetime(df_p_filtered['time'])

# Create a new column for the quarter start date
df_p_filtered.loc[:, 'quarter_start'] = df_p_filtered['time'].dt.to_period('Q').dt.start_time

# Group by latitude, longitude, and the quarter start date
# Calculate the mean temperature for each group
df_p_quarterly = df_p_filtered.groupby(['latitude', 'longitude', 'quarter_start']).agg({
    'precipitation_monthly-mean': 'mean'
})

# Rename columns for clarity
df_p_quarterly.rename(columns={'precipitation_monthly-mean': 'precipitation_quarterly_mean'}, inplace=True)
```
