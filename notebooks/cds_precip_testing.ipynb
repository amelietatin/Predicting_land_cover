```python
import xarray as xr

data_p= xr.open_dataset('../raw_data/precipitation_monthly-mean_bcc-csm1-1-m_rcp45_r1i1p1_1950-2100_v1.0.nc')
```

```python
df_p=data_p['precipitation_monthly-mean'].to_dataframe()
df_p
```

```python
df_p_reset = df_p.reset_index()
```

```python
import pandas as pd

df_p_reset['time'] = pd.to_datetime(df_p_reset['time'])

# Truncate the DataFrame to start from the year 2015
df_p_reset_truncated = df_p_reset[df_p_reset['time'].dt.year >= 2015]

# Reset the index if needed
df_p_reset_truncated.reset_index(inplace=True)
```

```python
df_p_reset_truncated
```

```python
min_lon= -31.169029141664772
max_lon = 32.611141887507934
min_lat = 27.746311186883922
max_lat = 69.04164438989706
```

```python
df_p_reset_truncated_filtered = df_p_reset_truncated[(df_p_reset_truncated["longitude"] >= min_lon) \
    & (df_p_reset_truncated["longitude"] <= max_lon) \
    & (df_p_reset_truncated["latitude"] >= min_lat) \
    & (df_p_reset_truncated["latitude"] <= max_lat)]
df_p_reset_truncated_filtered
```

```python
df_p_reset_truncated_filtered.set_index('index', inplace=True)
#df_p_reset_truncated_filtered.drop(columns=df_p_reset_truncated_filtered.columns[0], inplace=True)
df_p_reset_truncated_filtered
```

```python
df_p_reset_truncated_filtered.to_csv('../raw_data/precipitation_monthly_mean.csv')
```
