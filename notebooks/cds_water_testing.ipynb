```python
df_w=data_w['water-vapor-pressure'].to_dataframe()
df_w
```

```python
df_w_reset = df_w.reset_index()
```

```python
import pandas as pd

df_w_reset['time'] = pd.to_datetime(df_w_reset['time'])

# Truncate the DataFrame to start from the year 2015
df_w_reset_truncated = df_w_reset[df_w_reset['time'].dt.year >= 2015]

# Reset the index if needed
df_w_reset_truncated.reset_index(inplace=True)
```

```python
df_w_reset_truncated
```

```python
# min/max coordinate values for sample = 1000
# min_lon= -31.169029141664772
# max_lon = 32.611141887507934
# min_lat = 27.746311186883922
# max_lat = 69.04164438989706
```

```python
# min/max coordinate values for all PA
min_lon= -32.30906837949666
max_lon = 34.071555485388096
min_lat = 27.658764598717514
max_lat = 70.01981302035634
```

```python
df_w_filtered = df_w_reset_truncated[(df_w_reset_truncated["longitude"] >= min_lon) \
    & (df_w_reset_truncated["longitude"] <= max_lon) \
    & (df_w_reset_truncated["latitude"] >= min_lat) \
    & (df_w_reset_truncated["latitude"] <= max_lat)]
df_w_filtered
```

```python
df_w_filtered.set_index('index', inplace=True)
#df_w_reset_truncated_filtered.drop(columns=df_w_reset_truncated_filtered.columns[0], inplace=True)
df_w_filtered
```

```python
import pandas as pd

# Assuming df_t_reset_truncated is your initial DataFrame
# Ensure the time column is in datetime format
df_w_filtered.loc[:, 'time'] = pd.to_datetime(df_w_filtered['time'])

# Create a new column for the quarter start date
df_w_filtered.loc[:, 'quarter_start'] = df_w_filtered['time'].dt.to_period('Q').dt.start_time

# Group by latitude, longitude, and the quarter start date
# Calculate the mean temperature for each group
df_w_quarterly = df_w_filtered.groupby(['latitude', 'longitude', 'quarter_start']).agg({
    'water-vapor-pressure': 'mean'
})

# Rename columns for clarity
df_w_quarterly.rename(columns={'water-vapor-pressure': 'water-vapor-pressure_quarterly_mean'}, inplace=True)
```
