```python
df_c=data_c['cloud-cover'].to_dataframe()
df_c
```

```python
df_c_reset = df_c.reset_index()
```

```python
import pandas as pd

df_c_reset['time'] = pd.to_datetime(df_c_reset['time'])

# Truncate the DataFrame to start from the year 2015
df_c_reset_truncated = df_c_reset[df_c_reset['time'].dt.year >= 2015]

# Reset the index if needed
df_c_reset_truncated.reset_index(inplace=True)
```

```python
df_c_reset_truncated
```

```python
min_lon= -31.169029141664772
max_lon = 32.611141887507934
min_lat = 27.746311186883922
max_lat = 69.04164438989706
```

```python
df_c_reset_truncated_filtered = df_c_reset_truncated[(df_c_reset_truncated["longitude"] >= min_lon) \
    & (df_c_reset_truncated["longitude"] <= max_lon) \
    & (df_c_reset_truncated["latitude"] >= min_lat) \
    & (df_c_reset_truncated["latitude"] <= max_lat)]
df_c_reset_truncated_filtered
```

```python
df_c_reset_truncated_filtered.set_index('index', inplace=True)
#df_c_reset_truncated_filtered.drop(columns=df_c_reset_truncated_filtered.columns[0], inplace=True)
df_c_reset_truncated_filtered
```
