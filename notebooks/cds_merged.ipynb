### Merged CDS data for temperature and percipitation

```python
import pandas as pd

# Import the temperature CSV file
df_temp = pd.read_csv('../raw_data/temperature_monthly_mean.csv')

# Import the percipitation CSV file
df_precip = pd.read_csv('../raw_data/precipitation_monthly_mean.csv')

# Import the water CSV file
df_water = pd.read_csv('../raw_data/water-vapor-pressure.csv')

# Import the cloud CSV file
df_cloud = pd.read_csv('../raw_data/cloud-cover.csv')
```

```python
df_temp = pd.DataFrame(df_temp)
df_precip = pd.DataFrame(df_precip)
df_water = pd.DataFrame(df_water)
df_cloud = pd.DataFrame(df_cloud)
```

```python
import pandas as pd

# Merge df_temp and df_precip on index, latitude, longitude, and time
df_merged = pd.merge(df_temp, df_precip, on=['index', 'latitude', 'longitude', 'time'])

# Merge df_merged with df_water on index, latitude, longitude, and time
df_merged = pd.merge(df_merged, df_water, on=['index', 'latitude', 'longitude', 'time'])

# Merge df_merged with df_cloud on index, latitude, longitude, and time
df_merged = pd.merge(df_merged, df_cloud, on=['index', 'latitude', 'longitude', 'time'])
```

```python
df_merged.set_index('index', inplace=True)
df_merged
```

```python
df_merged['temperature_monthly-mean'] = (df_merged['temperature_monthly-mean'] - 273.15).round(2)
df_merged
```

### Sample with 10 PA including land cover proportions

```python
final_table_2015_10_pa = pd.read_csv('../raw_data/final_df_2015_10_pa.csv', sep=';')
final_table_2015_10_pa
```

```python
import pandas as pd

start_date = '2015-07-01'
#end_date = '2024-04-01'
end_date = '2015-12-01'

# for all years from 2015 to 2100:
# df_merged_all_years = df_merged[(df_merged['time'] >= start_date) & (df_merged['time'] <= end_date)]
# df_merged_all_years

# for 205 July - December (with sample = 10)
df_merged_2015 = df_merged[(df_merged['time'] >= start_date) & (df_merged['time'] <= end_date)]
df_merged_2015
```

```python
from scipy.spatial import distance_matrix
import pandas as pd
from datetime import datetime
import numpy as np

# Ensure datetime column is in datetime format
df_merged_2015['time'] = pd.to_datetime(df_merged_2015['time'])

# Define start and end dates
start_date = datetime(2015, 7, 1)
end_date = datetime(2015, 12, 1)

# Initialize the result DataFrame
result_2015_df = pd.DataFrame()

# Iterate through each month from start_date to end_date
current_date = start_date
while current_date <= end_date:
    # Filter df_merged_all_years for the current month
    df_current_month = df_merged_2015[
        (df_merged_2015['time'].dt.year == current_date.year) &
        (df_merged_2015['time'].dt.month == current_date.month)
    ]

    # Calculate distance matrix
    distances = distance_matrix(final_table_2015_10_pa[['lat', 'lon']], df_current_month[['latitude', 'longitude']])

    # Find the index of the nearest coordinates in df_current_month for each protected area
    nearest_idx = np.argmin(distances, axis=1)

    # Retrieve temperature and precipitation values for the nearest coordinates
    temperature_values = df_current_month.iloc[nearest_idx]['temperature_monthly-mean'].values
    precipitation_values = df_current_month.iloc[nearest_idx]['precipitation_monthly-mean'].values
    water_values = df_current_month.iloc[nearest_idx]['water-vapor-pressure'].values
    cloud_values = df_current_month.iloc[nearest_idx]['cloud-cover'].values

    # Create a temporary DataFrame to hold the results for this month
    temp_df = final_table_2015_10_pa.copy()
    temp_df['date'] = current_date
    temp_df['temperature_monthly_mean'] = temperature_values
    temp_df['precipitation_monthly_mean'] = precipitation_values
    temp_df['water_vapor_pressure'] = water_values
    temp_df['cloud_cover'] = cloud_values

    # Append the results for this month to the result DataFrame
    result_2015_df = pd.concat([result_2015_df, temp_df], ignore_index=True)

    # Move to the next month
    if current_date.month == 12:
        current_date = current_date.replace(year=current_date.year + 1, month=1, day=1)
    else:
        current_date = current_date.replace(month=current_date.month + 1, day=1)

# Now result_df contains temperature and precipitation values for each protected area for each month from January 2015 to December 2024
```
