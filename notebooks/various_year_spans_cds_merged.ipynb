### Merged CDS data for temperature, percipitation, cloud cover and water pressure

```python
import pandas as pd

# Import the temperature CSV file
df_temp = pd.read_csv('../raw_data/temperature_quarterly_mean.csv')

# Import the percipitation CSV file
df_precip = pd.read_csv('../raw_data/precipitation_quarterly_mean.csv')

# Import the water CSV file
df_water = pd.read_csv('../raw_data/water-vapor-pressure_quarterly.csv')

# Import the cloud CSV file
df_cloud = pd.read_csv('../raw_data/cloud-cover_quarterly.csv')
```

```python
df_temp = pd.DataFrame(df_temp)
df_precip = pd.DataFrame(df_precip)
df_water = pd.DataFrame(df_water)
df_cloud = pd.DataFrame(df_cloud)
```

```python
import pandas as pd

# Merge df_temp and df_precip on index, latitude, longitude, and time
df_merged = pd.merge(df_temp, df_precip, on=['latitude', 'longitude', 'quarter_start'])

# Merge df_merged with df_water on index, latitude, longitude, and time
df_merged = pd.merge(df_merged, df_water, on=['latitude', 'longitude', 'quarter_start'])

# Merge df_merged with df_cloud on index, latitude, longitude, and time
df_merged = pd.merge(df_merged, df_cloud, on=['latitude', 'longitude', 'quarter_start'])
```

### Sample for quarterly data from 2015-2024

```python
import geopandas as gp
import pandas as pd
```

```python
final_table_2015_2024_10_pa = pd.read_csv('../raw_data/final_lc_10_areas_2015_2024.csv', sep=';')
final_table_2015_2024_10_pa.head(50)
```

```python
import pandas as pd

start_date = '2015-07-01'
end_date = '2024-04-01'

# for all years from 2015 to 2024:
df_merged_2015_2024 = df_merged[(df_merged['quarter_start'] >= start_date) & (df_merged['quarter_start'] <= end_date)]
df_merged_2015_2024
```

```python
import pandas as pd
import numpy as np
from scipy.spatial import distance_matrix

# Ensure datetime columns are in datetime format
final_table_2015_2024_10_pa['date'] = pd.to_datetime(final_table_2015_2024_10_pa['date'])
df_merged_2015_2024['quarter_start'] = pd.to_datetime(df_merged_2015_2024['quarter_start'])

# Pivot the land cover data to get one row per SITECODE per quarter
df_landcover_pivot = final_table_2015_2024_10_pa.pivot_table(
    index=['SITECODE', 'date', 'lon', 'lat'],
    columns='LC',
    values='LC_PROPORTION'
).reset_index()

# Initialize the result DataFrame
result_df = pd.DataFrame()

# Iterate through each unique quarter
for quarter_start in df_landcover_pivot['date'].unique():
    # Filter data for the current quarter
    df_current_quarter = df_merged_2015_2024[df_merged_2015_2024['quarter_start'] == quarter_start]
    df_pa_current_quarter = df_landcover_pivot[df_landcover_pivot['date'] == quarter_start]

    # Calculate distance matrix between protected area coordinates and df_current_quarter coordinates
    distances = distance_matrix(df_pa_current_quarter[['lat', 'lon']], df_current_quarter[['latitude', 'longitude']])

    # Find the index of the nearest coordinates in df_current_quarter for each protected area
    nearest_idx = np.argmin(distances, axis=1)

    # Retrieve weather feature values for the nearest coordinates
    weather_values = df_current_quarter.iloc[nearest_idx][['temperature_quarterly_mean', 'precipitation_quarterly_mean', 'water-vapor-pressure_quarterly_mean', 'cloud-cover_quarterly_mean']].values

    # Add weather values to the current quarter's protected areas
    df_pa_current_quarter[['temperature_quarterly_mean', 'precipitation_quarterly_mean', 'water-vapor-pressure_quarterly_mean', 'cloud-cover_quarterly_mean']] = weather_values

    # Append the results for this quarter to the result DataFrame
    result_df = pd.concat([result_df, df_pa_current_quarter], ignore_index=True)

# Display the resulting DataFrame
```
